```
#https://www.kaggle.com/sudalairajkumar/daily-temperature-of-major-cities
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
#import plotly.graph_objs as go
#import plotly.offline as pyoff

#from google.colab import files
#uploaded = files.upload()
df = pd.read_csv('city_temperature.csv')
df.drop(['State'],axis='columns', inplace=True)

a = df[(df['Year'] == 2000)]
b = a[(df['Month'] == 8)]
c = b[(df['City'] == 'Osaka')]
x = c['Day']
y = c['AvgTemperature']
tt = sum(y)/31

fig = plt.figure()
fig.add_subplot(211)
plt.title('2000 - August - Osaka') 
plt.plot(x,y,'-ro')
plt.xlabel('Day Number')
plt.ylabel('Average_Temperature ')
plt.legend([u'Average= '+str(tt)])
# 29.29 C

#2
d = df[(df['Year'] == 2010)]
e = d[(df['Month'] == 8)]
f = e[(df['City'] == 'Osaka')]
i = f['Day']
j = f['AvgTemperature']
rr = sum(j)/31
fig.add_subplot(212)
plt.plot(i,j,'-go')
plt.xlabel('2010 - August - Osaka')
plt.ylabel('Average_Temperature ')
plt.legend([u'Average= '+str(rr)])
# 31.11 C
plt.show()

# Years array from 2000 to 2010
years = np.linspace(2000,2010, endpoint=True, num=25)

# Toss average of 2020 due to being partial year at time of making. df2 will be used for annual average analysis
df2 = df[df["Year"] < 2020]
e2 = df2[(df['Month'] ==8 )]
f2 = e2[(df['City'] == 'Osaka' )]
# Toss temperatures below 10 and above 125
df2 = f2[f2["AvgTemperature"] > 10]
df2 = f2[f2["AvgTemperature"] < 125]

# Group data by year and find mean
means = df2.groupby("Year")["AvgTemperature"].mean()
a = means

# Plot data with average line
plt.figure(figsize=(16,8))
plt.plot(a)
plt.xlabel("Year")
plt.ylabel("Average Japan Temperature, Fahrenheit")
plt.title("Average Japan Temperatures by Year (August ONLY)")
arr = pd.Series(a).array
m, b = np.polyfit(years, arr, 1)
plt.plot(years, m*years + b)
plt.legend(['Average Temperature by Year', 'Average Japan'])



```
